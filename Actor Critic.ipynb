```python
import matplotlib.pyplot as plt


def plot_running_reward(running_reward_list, reward_per_ep_list, varname, fname):
    try:
        plt.close()
    except:
        pass
    plt.plot(range(len(running_reward_list)), reward_per_ep_list)
    plt.plot(range(len(running_reward_list)), running_reward_list)
    plt.xlabel('Episode')
    plt.ylabel('Running '+varname)
    plt.title('Running '+varname+' over Training Episodes of'+fname)
    fname = varname+" vs " + str(len(running_reward_list)) + " Epochs" + str(fname)
    plt.savefig("plots/" + fname + '.png')
    plt.show()
    plt.pause(0.001)  # pause a bit so that plots are updated
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import gym
import numpy as np
from gym.wrappers import AtariPreprocessing, FrameStack
import torch.nn.functional as F

from play import *


debug = False
def printf(*args):
    if debug:
        print(" ".join(map(str,args)))
        
# Define the Actor and Critic Networks
class ActorCritic(nn.Module):
    def __init__(self, input_size, output_size, hidden_size):
        super(ActorCritic, self).__init__()
        
        #self.n_actions = out_size = n_actions
        
        
        self.conv1 = nn.Conv2d(4, 32, kernel_size=8, stride=4)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=4, stride=2)
        self.conv3 = nn.Conv2d(64, 64, kernel_size=3, stride=1)
        self.fc1 = nn.Linear(7*7*64, 512)
        
        self.action_layer = nn.Linear(512, output_size)
        self.value_layer = nn.Linear(512, 1)
        
        self.actor = nn.Sequential(
            nn.Linear(512, output_size),
            nn.Softmax(dim=-1)
        )
        self.critic = nn.Sequential(
            nn.Linear(512, 1)
        )

    def forward(self, x):
        x = x.clamp(0, 255).to(torch.uint8)
        assert x.dtype == torch.uint8, "The model expects states of type ByteTensor"
        x = x.float().div(255)
        
        
        x = F.relu(self.conv1(x))
        printf("After conv1", x.shape)
        x = F.relu(self.conv2(x))
        printf("After conv2", x.shape)
        x = F.relu(self.conv3(x))
        printf("After conv3", x.shape)
        
        x = x.view(x.size(0), -1)
        printf("After reshaping", x.shape)
        
        x = F.relu(self.fc1(x))
        printf("After fc1", x.shape)
        
        
        policy = self.actor(x)
        printf("After actor", policy.shape)
        
        value = self.critic(x)
        printf("After value", value.shape)
        
        return policy, value

class ActorCriticPre(nn.Module):
    def __init__(self, input_size, output_size, hidden_size):
        super(ActorCriticPre, self).__init__()
        
        #self.n_actions = out_size = n_actions
        
        
        self.__features = pretrained_model._AtariNet__features
        for param in self.__features.parameters():
            param.requires_grad = False
        self.fc1 = nn.Linear(7*7*64, 512)
        
        self.action_layer = nn.Linear(512, output_size)
        self.value_layer = nn.Linear(512, 1)
        
        self.actor = nn.Sequential(
            nn.Linear(512, output_size),
            nn.Softmax(dim=-1)
        )
        self.critic = nn.Sequential(
            nn.Linear(512, 1)
        )

    def forward(self, x):
        x = x.clamp(0, 255).to(torch.uint8)
        assert x.dtype == torch.uint8, "The model expects states of type ByteTensor"
        x = x.float().div(255)
        
        
        x = self.__features(x)
        
        x = x.view(x.size(0), -1)
        printf("After reshaping", x.shape)
        
        x = F.relu(self.fc1(x))
        printf("After fc1", x.shape)
        
        
        policy = self.actor(x)
        printf("After actor", policy.shape)
        
        value = self.critic(x)
        printf("After value", value.shape)
        
        return policy, value

    
# Define the Actor-Critic Agent
class ACAgent:
    def __init__(self, env, device, lr=0.01, gamma=0.99, hidden_size=256):
        self.env = env
        self.gamma = gamma
        self.model = ActorCriticPre(env.observation_space.shape[0], env.action_space.n, hidden_size).to(device)
        self.optimizer = optim.Adam(self.model.parameters(), lr=lr)
        self.device = device

    def act(self, state):
        #state = torch.FloatTensor(state).to(self.device)
        policy, _ = self.model(state)
        action = torch.distributions.Categorical(policy).sample()
        return action.item()

    def update(self, rewards, log_probs, values):
        # Calculate the returns
        returns = []
        G = 0
        for r in rewards[::-1]:
            G = r + self.gamma * G
            returns.insert(0, G)
        returns = torch.tensor(returns).to(self.device)

        # Calculate the advantages
        advantages = returns - values

        # Calculate the policy loss and value loss
        policy_loss = (-log_probs * advantages.detach()).mean()
        value_loss = nn.MSELoss()(values, returns.view(-1, 1, 1))

        # Backpropagate the total loss and update the parameters
        loss = policy_loss + value_loss
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()

    def train(self, num_episodes):
        device = self.device
        fname = "NonTrained"
        running_reward_list = []
        reward_per_ep_list = []
        sumi = 0
        
        sumi_duri = 0
        running_duration_list = []
        duration_per_ep_list = []
        for episode in range(num_episodes):
            print(f"In {episode}th episode--------------> ")
            state, _= self.env.reset()
            state = torch.tensor(state, dtype=torch.float32, device=device).unsqueeze(0)
            
            rewards = []
            log_probs = []
            values = []

            while True:
                # Choose an action
                action = self.act(state)

                # Take the action and observe the next state and reward
                next_state, reward, done, _, _ = self.env.step(action)
                next_state = torch.tensor(next_state, dtype=torch.float32, device=device).unsqueeze(0)

                # Save the reward
                rewards.append(reward)

                # Calculate the log probability and value of the current state
                #state = torch.FloatTensor(state)
                policy, value = self.model(state)
                log_prob = torch.log(policy[0][action])

                # Save the log probability and value
                log_probs.append(log_prob)
                values.append(value)

                if done:
                    # Update the actor-critic network
                    self.update(rewards, torch.stack(log_probs), torch.stack(values))
                    break

                state = next_state
            
            epReward = np.sum(rewards)
            reward_per_ep_list.append(epReward)
            sumi += epReward
            running_reward_list.append(sumi/(episode+1))

            epDuration = len(rewards)
            duration_per_ep_list.append(epDuration)
            sumi_duri += epDuration
            running_duration_list.append(sumi_duri/(episode+1))
            
            

            if episode%50 == 0:
                plot_running_reward(running_reward_list, reward_per_ep_list, "Reward", fname)
                plot_running_reward(running_duration_list, duration_per_ep_list, "Duration", fname)
                torch.save(self.model.state_dict(), './preTrainedMODELS/Breakout_{}_ep{}.pth'.format(fname, episode))
                print('Episode {}\tlength: {}\treward: {}'.format(episode, epDuration, epReward))
                



            
# Run the Actor-Critic Agent on the CartPole-v1 environment
env = gym.make("BreakoutNoFrameskip-v4", render_mode = "rgb_array")
env = AtariPreprocessing(env, scale_obs=False)
env = FrameStack(env, num_stack=4)
record_every = 100
#record video for every even episodes
env = gym.wrappers.RecordVideo(env, 'video', episode_trigger = lambda x: x % record_every == 0)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using",device)

def _load_checkpoint(fpath, device="cpu"):
    fpath = Path(fpath)
    with fpath.open("rb") as file:
        with GzipFile(fileobj=file) as inflated:
            return torch.load(inflated, map_location=device)
        
path = "models/DQN_modern/Breakout/2/model_50000000.gz"

pretrained_model = AtariNet(env.action_space.n, distributional="C51_" in path)
ckpt = _load_checkpoint(path)
pretrained_model.load_state_dict(ckpt["estimator_state"])

agent = ACAgent(env, device)
agent.train(num_episodes=1000)
```
